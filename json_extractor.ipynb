```python
# delete 10.7910_DVN_TZ4C1Q.json from Big Collection
# runs in about 5 minutes, returns about 232137 rows
import json
from glob import glob
import pandas as pd


def main():

    paths = glob('C:/Users/Premkumar/Desktop/College/VA_final_presentation/BigCollection/BigCollection/*.json') # change to relevant directory
    big_dict = {'title': [], 'date': [], 'url': [],'distributorname': [],'distributionDate': [],'distributorAbbreviation': [] , 'keyword': []}
    
    for path in paths:
        data = _extract_data_from_file(path)
        for keyword in data['keywords']:
            big_dict['title'].append(data['title'])
            big_dict['date'].append(pd.to_datetime(data['date']))
            big_dict['url'].append(data['url'])
            big_dict['distributorname'].append(data['dName'])
            big_dict['distributionDate'].append(data['dDate'])
            big_dict['distributorAbbreviation'].append(data['dAbbrev'])
            
            big_dict['keyword'].append(keyword.lower())

    df = pd.DataFrame.from_dict(big_dict).dropna()
    df.to_csv(r'C:/Users/Premkumar/Desktop/College/VA_final_presentation/CSV_file/big_collection_final.csv', index = None, header=True) # change
    
    
def _extract_data_from_file(filename):

    with open(filename, 'r', encoding="utf8") as f:
        obj = json.load(f)

    url = obj['data']['persistentUrl']
    date = obj['data']['publicationDate']
    citation = obj['data']['latestVersion']['metadataBlocks']['citation']['fields']
    keywords = []
    dName=''
    dAbbrev=''
    dAffl=''
    dDate=''
    for field in citation:
        if field['typeName'] == 'title':
            title = field['value']
        if field['typeName'] == 'distributor':
            for d in field['value']:
                if 'distributorName' in d:
                    dName=d['distributorName']['value']
                if 'distributorAbbreviation' in d:
                    dAbbrev=d['distributorAbbreviation']['value']
                if 'distributorAffiliation' in d:
                    dAffl=d['distributorAffiliation']['value']
        if field['typeName'] == 'distributionDate':
            dDate=field['value']
            
            
            #print(field['value']['distributorName']['typeName'])
#             if field['value']['distributorName']['typeName'] == 'distributorName':
#                 # dName=
#                 print(field['distributorName']['value'])
#             if field['value']['distributorAbbreviation']['typeName'] == 'distributorAbbreviation':
#                 dAbbrev=field['value']['distributorAbbreviation']['value']
        
        
        if field['typeName'] == 'keyword':
            for k in field['value']:
                if 'keywordValue' in k:
                    keywords.append(k['keywordValue']['value'])
    
    #print('title: ', title, 'date: ', date, 'url: ', url, 'keywords: ', keywords)
   # print(dName,'    dabbrevation: ',dAbbrev)

    return {'title': title, 'date': date, 'url': url,'dName': dName ,'dAbbrev': dAbbrev,'dDate': dDate, 'keywords': keywords}


main()
```
